```python
my_numpy3 = np.array([ [1,2], [4,5] ])
my_numpy4 = np.array([ [10, 12], [9,26] ])
print('my_numpy3: ', my_numpy3)
print('my_numpy4: ', my_numpy4)
```

```python
# Array attributes
print("my_numpy4.ndim: ", my_numpy4.ndim)  # ndim : Represents the number of dimensions or "rank" of the array.

print("my_numpy4.shape: ", my_numpy4.shape)  # shape : Returns a tuple indicating the number of rows and columns in the array.

print("my_numpy4.size: ", my_numpy4.size) # size: Provides the total number of elements in the array.
print(type(my_numpy4.shape))
```

```python
print(my_numpy4[1]) # Accessing a row
print(my_numpy4[1,1]) # Accessing an element
```

```python
print(my_numpy4[:, 1])       # Accessing a column (2nd column)
print(my_numpy4[1])       # Accessing a row
print(my_numpy4[1, :])       # Accessing a row - same as above
```
