```python
my_dataDictionary = {'d1': [1, 2, 2], 'd2': [11,12,2.6], 'd3': [111,121, 'data 3'], 'd4': [1111,1211, 'data 4']}
my_dataFrame = pd.DataFrame(my_dataDictionary)
my_dataFrame.head()
```

```python
print(my_dataFrame.iloc[1,1]) # Location based indexing can only have integer
print(my_dataFrame.iloc[2,1])
print(my_dataFrame.loc[0,'d4']) # Location based on column name
```

```python
print(my_dataFrame.iloc[0:2,1])
```

```python
print(my_dataFrame.loc[0:2,'d4'])
```

```python
print(my_dataFrame.loc[0:1,'d4'])
```

```python
print(my_dataFrame.dtypes)
print(my_dataFrame['d1'].dtype) # datatype of a single column
```

```python
unique = my_dataFrame['d1'].unique()
print(unique)
```

```python
my_dataFrame2 = my_dataFrame['d2'] > 10
print(my_dataFrame2)
```

```python
my_dataFrame2 = my_dataFrame[my_dataFrame['d2'] > 10]
print(my_dataFrame2)
```
