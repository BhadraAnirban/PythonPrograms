```python
mydata = ['red', 'green', 'blue']
for data in range(0,3):
    mydata[data] = mydata[data] + '2'
print(mydata)
```

```python
mydata = ['red', 'green', 'blue']
for data in mydata:
    data = data+'2'
    print(data)
print(mydata)
```

```python
mydata = ['red', 'green', 'blue']
newData = []
for data in mydata:
    newData.append(data + '2')
print(mydata)
print(newData)
```

```python

mydata = ['red', 'green', 'blue']
mydata = [color + '3' for color in mydata]
print(mydata)
```

```python
mydata = ['red', 'green', 'blue']
for dataIndex in range(0,len(mydata)):
    mydata[dataIndex] = mydata[dataIndex] + '2'
print(mydata)
```

```python
mydata = ['red', 'green', 'blue']
for data in range(len(mydata)):
    mydata[data] = mydata[data] + '2'
print(mydata)
```

```python

mydata = ['red', 'green', 'blue']
for i, data in enumerate(mydata):
    mydata[i] = data + '2'
print(mydata)
```

```python

mydata = ['red', 'green', 'blue']
mydata = [color + '2' for color in mydata] # create a new list and assign it to mydata
print(mydata)
```

```python
mydata = ['red', 'green', 'blue', 'yellow', 'black']
newData = []

i = 0
while(mydata[i] != 'yellow'):
    newData.append(mydata[i])
    i = i+1

print(newData)
```
