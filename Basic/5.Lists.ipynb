### Lists are in ordered sequence written in comma seperated with in square brackets []
#### Lists are mutable. Means we can do => list_var[2] = 2.26

```python
list_var = ['BBM', 2, 'PMKJ', 26]
```

```python
print(len(list_var))
print(list_var[0])
print(list_var[1])
print(list_var[2])

```

```python
list_var[1] = list_var[1]*2
print(list_var[1])
```

```python
list_var[3] = 26.2
print(list_var[3])
```

### split operation convert string into list (not tuple)

```python
new_string = "Pranabananda Maharaj Ki Jay"
```

```python
new_list = new_string.split()
```

```python
print(new_list)
```

```python
duplicate_list = new_list
```

```python
print(duplicate_list)
duplicate_list[1] = 2
print(new_list)
```

#### Clone the list using [:]

```python
clone_list = new_list[:]
clone_list[1] = 'Maharaj'
print(clone_list)
print(new_list)
```

### The copy() method is used to create a shallow copy of a list.
### The count() method is used to count the number of occurrences of a specific element in a list in Python.
### The del statement is used to remove an element from list. It removes the element at the specified index.

```python
my_list = [1, 2, 2, 3, 4, 2, 5, 2] 
count = my_list.count(2); print(count) 
my_list = [10, 20, 30, 40, 50] 
print(my_list)
del my_list[2] # Removes the element at index 2 print(my_list) 
print(my_list)
```

### The insert() method is used to insert an element.

```python
my_list = [1, 2, 3, 4, 5] 
my_list.insert(2, 6) #list_name.insert(index, element
print(my_list)
```

### pop and remove method

```python
my_list = [10, 20, 30, 40, 50] 
removed_element = my_list.pop(2) # Removes and returns the element at index 2 
print(removed_element) 
print(my_list) 
```

```python
my_list = [10, 20, 30, 40, 50] 
my_list.remove(30) # Removes the element 30 
print(my_list) 
```

### Slicing

```python
my_list = [1, 2, 3, 4, 5] 
print(my_list[1:4]) 
# Output: [2, 3, 4] (elements from index 1 to 3)

print(my_list[:3]) 
# Output: [1, 2, 3] (elements from the beginning up to index 2) 

print(my_list[2:]) 
# Output: [3, 4, 5] (elements from index 2 to the end) 

print(my_list[::2]) 
# Output: [1, 3, 5] (every second element)
```
