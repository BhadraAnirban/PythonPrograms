```python
class Circle(object):
    def __init__(self, radius, color):
        self.radius = radius;
        self.color = color;
    def getArea(self):
        return 3.14*self.radius*self.radius;
```

```python
mycircle = Circle(26, 'red');
print(f"The area is: {mycircle.getArea()}");
print("The color is: ", mycircle.color)
```

```python
class Rectangle(object):
    color = 'yellow';
    def __init__(self, width, hight, color):
        self.width = width;
        self.hight = hight;
        self.color = color;
    def getArea(self):
        return 3.14*self.width* self.hight;

    def getColor(self):
        return self.color;
```

```python
rect = Rectangle(2, 4, 'blue')
print(rect.getArea())
print(rect.getColor())
```

```python
rect.color = 'green';
print(rect.getColor())
```

#### Private (__double underscore)

```python
class Rectangle2(object):
    color = 'yellow';
    __weight = 20;  # Private (__double underscore)
    def __init__(self, width, hight):
        self.width = width;
        self.hight = hight;
        print(f"In Constructor: {self.width}")
    
    def setColor(self, color):
        self.color = color;
        return None;

    def getColor(self):
        return self.color;
```

```python
rect2 = Rectangle2(2, 4);
print(rect2.getColor());
rect2.color = input('Enter color: ')
print(rect2.getColor());
```

```python
print(rect2.__weight) # Direct access will fail
```

```python
print(rect2._Rectangle2__weight) # Access using name mangling
```
