```python
def fun1(x):
    try:
        print(x/2)
    except:
        print(F"{x} is a: issue")
    else:
        print(F"{x} is a: success")
```

```python
fun1(10)
fun1(1)
fun1('pmkj')
```

```python
def fun2(x):
    try:
        if(x == 0):
            raise ValueError("x cannot be zero")
        if(x == 1):
            raise Exception("Something went wrong")
        print(x/2)
    except Exception as ex:
        print(f"{x} is a: {ex}") # str(ex) -> ex.message
    else:
        print(F"{x} is a: success")
```

```python
fun2(10)
fun2(0)
fun2(1)
fun2('pmkj')
```

```python
def fun3(x):
    if(x == 0):
        raise ValueError("x cannot be zero")
    elif(x == 1):
        raise Exception(f"Something went wrong with {x}")
    print(x/2)
```

```python
fun3(10)
fun3(0)
fun3(1)
fun3('pmkj')
```

```python
try:
    fun3(17);
    fun3(0);
    fun3(1);
    fun3(11);
    fun3('pmkj');
except Exception as ex:
    print(ex)
```
